```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/transactions_preprocessed.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(x='fraud', data=df, palette=['#FF9999', '#66B2FF'])
plt.title('Distribuição de Fraudes', fontsize=16)
plt.xlabel('Fraude', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
plt.xticks([0, 1], ['Não Fraudulenta', 'Fraudulenta'])
plt.show()

numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor das Correlações', fontsize=16)
plt.show()
```
